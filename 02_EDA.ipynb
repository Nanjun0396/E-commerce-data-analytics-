## Overview
This notebook explores the cleaned UK e-commerce dataset to answer key business questions.
We analyze revenue patterns, customer behavior, product performance, and seasonal trends.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\Guganathan\Desktop\ecommerce-data-analysis\data\data_cleaned.csv')
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print("Data loaded! Shape:", df.shape)
```

## Analysis 1 — Revenue by Country
Which markets generate the most revenue for the business?

```python
# Which countries generate the most revenue?
revenue_by_country = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False)
print(revenue_by_country.head(10))
```

## Analysis 2 — Top Products by Revenue
Which products drive the most revenue vs highest sales volume?

```python
# Which products generate the most revenue?
revenue_by_product = df.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False)
print(revenue_by_product.head(10))
```

## Analysis 3 — Monthly Revenue Trends
How does revenue change throughout the year? Are there seasonal patterns?

```python
# How does revenue trend across months?
monthly_revenue = df.groupby(['Year', 'Month'])['TotalPrice'].sum()
print(monthly_revenue)
```

```python
# Which products are ordered most often?
orders_by_product = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False)
print(orders_by_product.head(10))
```

```python
# How does revenue change month by month?
monthly_revenue = df.groupby(['Year', 'Month'])['TotalPrice'].sum()
print(monthly_revenue)
```

## Analysis 4 — Revenue by Day of Week
Which days generate the most sales? This helps identify the nature of the business.

```python
# Which days of the week have highest sales?
revenue_by_day = df.groupby('DayOfWeek')['TotalPrice'].sum().sort_values(ascending=False)
print(revenue_by_day)
```

## Analysis 5 — Top Customers
Who are the highest value customers and does the Pareto principle apply?

```python
# Who are the top 10 customers by revenue?
top_customers = df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False)
print(top_customers.head(10))
```

```python
# What percentage of revenue comes from top 20% of customers?
total_customers = df['CustomerID'].nunique()
top_20_percent = int(total_customers * 0.2)
top_customers_revenue = df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False)

top_20_revenue = top_customers_revenue.head(top_20_percent).sum()
total_revenue = top_customers_revenue.sum()

print(f"Total customers: {total_customers}")
print(f"Top 20% customers: {top_20_percent}")
print(f"Revenue from top 20%: £{round(top_20_revenue, 2)}")
print(f"Percentage of total revenue: {round(top_20_revenue/total_revenue*100, 1)}%")
```
