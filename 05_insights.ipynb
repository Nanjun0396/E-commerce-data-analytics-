## Overview
This notebook summarizes all key business insights discovered through EDA and SQL analysis.
Each insight includes a business recommendation backed by data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\Guganathan\Desktop\ecommerce-data-analysis\data\data_cleaned.csv')
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

print("Data loaded! Shape:", df.shape)
```

## Business Overview
High level KPIs summarizing the scale of the business analyzed.

```python
# Business Insights Summary
total_revenue = df['TotalPrice'].sum()
total_customers = df['CustomerID'].nunique()
total_orders = df['InvoiceNo'].nunique()
total_products = df['Description'].nunique()
avg_order_value = df.groupby('InvoiceNo')['TotalPrice'].sum().mean()

print("=" * 50)
print("ECOMMERCE BUSINESS INSIGHTS SUMMARY")
print("=" * 50)
print(f"Total Revenue:        £{round(total_revenue, 2):,}")
print(f"Total Customers:      {total_customers:,}")
print(f"Total Orders:         {total_orders:,}")
print(f"Total Products:       {total_products:,}")
print(f"Avg Order Value:      £{round(avg_order_value, 2)}")
print("=" * 50)
```

## Key Business Insights
Four data-driven insights with actionable business recommendations.

```python
# Key Business Insights
print("=" * 50)
print("TOP BUSINESS INSIGHTS")
print("=" * 50)

# Insight 1 - UK Dominance
uk_revenue = df[df['Country'] == 'United Kingdom']['TotalPrice'].sum()
uk_pct = round(uk_revenue / total_revenue * 100, 1)
print(f"\n1. UK MARKET DOMINANCE")
print(f"   UK generates {uk_pct}% of total revenue (£{round(uk_revenue,2):,})")
print(f"   Opportunity: Expand marketing in Netherlands & Australia")
print(f"   where average order values are highest (£122 and £117)")

# Insight 2 - Seasonality
monthly = df.groupby('Month')['TotalPrice'].sum()
peak_month = monthly.idxmax()
low_month = monthly.idxmin()
print(f"\n2. STRONG SEASONAL PATTERN")
print(f"   Peak month: Month {peak_month} (November) - £{round(monthly[peak_month],2):,}")
print(f"   Lowest month: Month {low_month} (February) - £{round(monthly[low_month],2):,}")
print(f"   November revenue is {round(monthly[peak_month]/monthly[low_month],1)}x higher than February")

# Insight 3 - 80/20 Rule
total_custs = df['CustomerID'].nunique()
top_20 = int(total_custs * 0.2)
top_20_revenue = df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False).head(top_20).sum()
top_20_pct = round(top_20_revenue / total_revenue * 100, 1)
print(f"\n3. PARETO PRINCIPLE CONFIRMED")
print(f"   Top 20% of customers ({top_20}) generate {top_20_pct}% of revenue")
print(f"   Bottom 80% of customers generate only {100-top_20_pct}% of revenue")
print(f"   Focus: Retain and upsell top customers")

# Insight 4 - One time buyers
one_time = df.groupby('CustomerID')['InvoiceNo'].nunique()
one_time_pct = round((one_time == 1).sum() / total_custs * 100, 1)
print(f"\n4. CUSTOMER RETENTION OPPORTUNITY")
print(f"   {one_time_pct}% of customers only bought once (1,505 customers)")
print(f"   10+ order customers spend 35x more than one-time buyers")
print(f"   Focus: Win-back campaigns for one-time buyers")

print("\n" + "=" * 50)
```

```python
# Fix insight 3 display
bottom_80_pct = round(100 - top_20_pct, 1)
print(f"Bottom 80% of customers generate only {bottom_80_pct}% of revenue")
```
